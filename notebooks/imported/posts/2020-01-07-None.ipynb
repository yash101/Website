## Strengths and Weaknesses of Electricity Sources

### Solar Photovoltaic

Solar photovoltaic generation uses panels of photovoltaic cells to directly
generate electricity from the light that hits it. These panels are commonly
known as solar panels. [Find more information about PV systems here
(Wikipedia).](https://en.wikipedia.org/wiki/Photovoltaic_system)

#### Strengths:

  * Easy to install
  * Works off-grid well
  * No carbon emissions
  * Solid state - no potential for mechanical failures

#### Weaknesses:

  * Generally, significant energy requirements for production -> CO2
  * Uses scarce resources
  * Production results in many toxic byproducts
  * No way to store excess energy easily
  * Requires sunlight:
    * Less sunlight at higher/lower latitudes
    * Only works in the day
    * Requires good, sunny weather
    * Amount of energy generated varys by season
  * Requires tracking for optimal operation (nullifies the solid state strength)

### Solar Thermal (Concentrated Solar Power)

Unlike solar photovoltaic generation systems, which convert light energy
directly to an electric potential, solar thermal plants concentrate sunlight
and heat a medium, which is used to generate a different type of generator,
such as a steam engine. For more information about solar thermal generators, I
like these resources: [U.S. Energy Information
Administration](https://www.eia.gov/energyexplained/solar/solar-thermal-power-
plants.php), [Wikipedia
page](https://en.wikipedia.org/wiki/Solar_thermal_energy), [U.S. Department of
Energy / Office of Efficiency and Renewable
Energy](https://www.energy.gov/eere/solar/concentrating-solar-power).

#### Strengths:

  * No carbon emissions
  * Less expensive than PV plants
  * Thermal storage is easy and reliable
  * A backup thermal generator can be added onto the thermal network to provide power when sunlight is not available
  * Can be retrofitted to existing steam and thermal plants (such as coal plants) - this also can potentially reduce costs

#### Weaknesses:

  * Tons of moving parts
    * The moving thermal transfer medium can degrade over time, and pipes can be damaged over time from the higher temperatures
    * The electricity generation method likely has moving parts
  * Can poses danger to wildlife
    * Birds are instantly killed or blinded if they fly through concentrated light beams
