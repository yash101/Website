### CPU vs GPU

The easiest way to understand the difference between a CPU and a GPU is through an analogy.

**A CPU is a lot like a semitruck**. It's capable of moving a small number of goods from point A to point Z, stopping along the way loading and unloading cargo at different points.



<br />

**A GPU is much more akin to a container ship**. Within one trip of the container ship, it can transport multiple containers of cargo. Cargo ships can transport containers at a fraction of the cost of a semitruck, but they have limitations:

1. Stops are expensive
2. Packaging and unpackaging the containers to and from the ship is harder
3. It requires much more skill and delicate placement of containers on the container ship to efficiently load and unload containers at future destinations



**But here's the kicker:** transporting a container on a cargo ship costs a fraction as much as it does to transport it via semitruck or even rail!

Now, if we imagine each of the containers being transported in the analogy as "data", it helps paint a picture of what GPUs and CPUs excel at.

### Why GPUs Were Ever Invented

Why do we even need GPUs? Aren't CPUs powerful enough for graphics processing?

GPUs were created because **graphics processing is hard**.

A 4k display is 3840 pixels wide and 2160 pixels tall. There are three channels (red/R, green/G, blue/B). At 60 FPS, how many pixels do we need to render?

$$3840\text{ px}\times2160\text{ px}\times3\text{ ch} = 24\ 883\ 200\frac{\text{px}}{\text{frame}}$$

Oh, and we are processing 60 frames per second.

$$24\ 883\ 200\ \frac{\text{pixels}}{\text{frame}} \times 60\ \frac{\text{frame}}{s}=1\ 492\ 992\ 000\ \frac{\text{px}}{s}$$

That's nearly $1.5$ billion pixels we need to render each and every second. And 4k60 isn't even bleeding edge anymore. Try 8k120!
