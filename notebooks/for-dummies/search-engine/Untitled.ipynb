```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont

# Generate Zipfian data
N = 50  # Number of words
ranks = np.arange(1, N + 1)
s = 1.1  # Zipf exponent
frequencies = (ranks ** -s) / np.sum(ranks ** -s) * 1000  # Normalize

# Create figure
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(ranks, frequencies, marker="o", linestyle="-", color="royalblue", label="Zipf's Law")

# Annotate common vs. rare words
common_words = ["the", "and", "of", "to", "in", "a", "is", "that", "it", "on"]
rare_words = ["serendipity", "ephemeral", "quixotic", "zeitgeist", "callipygian"]

for i, word in enumerate(common_words):
    ax.text(ranks[i], frequencies[i] + 10, word, fontsize=10, ha="right", color="black")

for i, word in enumerate(rare_words):
    idx = N - i - 1
    ax.text(ranks[idx], frequencies[idx] + 10, word, fontsize=10, ha="left", color="darkred")

ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("Word Rank (log scale)")
ax.set_ylabel("Word Frequency (log scale)")
ax.set_title("Zipf's Law in Language")
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
V0 = 1500              # Initial voltage (V)
R = 50                 # Load resistance (ohms)
C = 133e-6             # Capacitance (F)
tau = R * C          # Time constant (s)
T1 = 5e-3            # Phase 1 duration (5 ms)
T2 = 5e-3            # Phase 2 duration (5 ms)

# Time arrays for each phase
t_phase1 = np.linspace(0, T1, 100)
t_phase2 = np.linspace(T1, T1+T2, 100)

# Phase 1: Positive pulse
V_phase1 = V0 * np.exp(-t_phase1/tau)
I_phase1 = V_phase1 / R

# Phase 2: Negative pulse
V_phase2 = -V0 * np.exp(-(t_phase2-T1)/tau)
I_phase2 = V_phase2 / R

# Combine phases
t = np.concatenate((t_phase1, t_phase2))
V = np.concatenate((V_phase1, V_phase2))
I = np.concatenate((I_phase1, I_phase2))

# Plotting
plt.figure(figsize=(10, 6))

plt.subplot(2,1,1)
plt.plot(t*1e3, V)
plt.xlabel('Time (ms)')
plt.ylabel('Voltage (V)')
plt.title('Biphasic Voltage Waveform')
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(t*1e3, I)
plt.xlabel('Time (ms)')
plt.ylabel('Current (A)')
plt.title('Biphasic Current Waveform')
plt.grid(True)

plt.tight_layout()
plt.show()
```
